### Базовая очистка названий колонок

```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(",", "", regex=False)
    .str.replace("  ", " ", regex=False)
)
```

```python
df.columns
```

### Переименование колонок в рабочий формат

```python
rename_map = {
    "Отметка времени": "timestamp",
    "Age group": "age_group",
    "Gender": "gender",
    "Status": "status",

    "Average night sleep duration (past 2 weeks)": "sleep_duration",
    "Overall night sleep quality (past 2 weeks)": "sleep_quality",
    "Average number of awakenings per night (past 2 weeks)": "night_awakenings",

    "How often do you take daytime naps? (past 2 weeks)": "nap_frequency",
    "Average duration of a daytime nap": "nap_duration",
    "Typical time of daytime nap": "nap_time",
    "Main reason for daytime naps (choose one)": "nap_reason",

    "Self-reported productivity — today": "productivity_today",
    "Average productivity over the past 7 days": "productivity_week",
    "Task completion compared to your usual level": "task_completion",

    "Current stress level": "stress_current",
    "Stress level one week ago": "stress_week",
    "Stress level one month ago": "stress_month",

    "Current anxiety level": "anxiety_current",
    "Anxiety level one week ago": "anxiety_week",
    "Anxiety level one month ago": "anxiety_month",

    "Current fatigue level": "fatigue_current",
    "Fatigue level one week ago": "fatigue_week",
    "Fatigue level one month ago": "fatigue_month",

    "Workouts per week (past month)": "workouts_per_week",
    "Average duration of workout": "workout_duration",
    "Main type of physical activity": "activity_type",

    "Overall diet quality (past month)": "diet_quality",
    "How often do you skip meals?": "skip_meals",
    "Fast food consumption per week": "fast_food_per_week",

    "Average daily caffeine intake": "caffeine_daily",
    "Latest usual time of caffeine consumption": "caffeine_last_time",

    "Daily screen time": "screen_time_daily",
    "Screen time after 22:00": "screen_time_late",
    "Impact of screen use on falling asleep": "screen_sleep_impact",

    "Overall life satisfaction — current": "life_satisfaction_current",
    "Life satisfaction one month ago": "life_satisfaction_month",

    "Work/studies": "satisfaction_work",
    "Social relationships": "satisfaction_social",
    "Physical health": "satisfaction_health",
    "Emotional well-being": "satisfaction_emotional"
}

df = df.rename(columns=rename_map)
```

```python
df.head()
```

### Приведение типов данных

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")
```

```python
numeric_cols = [
    "sleep_quality",
    "night_awakenings",
    "productivity_today",
    "productivity_week",
    "stress_current", "stress_week", "stress_month",
    "anxiety_current", "anxiety_week", "anxiety_month",
    "fatigue_current", "fatigue_week", "fatigue_month",
    "diet_quality",
    "life_satisfaction_current", "life_satisfaction_month",
    "satisfaction_work", "satisfaction_social",
    "satisfaction_health", "satisfaction_emotional"
]

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

### Логическая очистка значений

```python
scale_cols = numeric_cols

for col in scale_cols:
    df.loc[(df[col] < 0) | (df[col] > 10), col] = np.nan 
```

## Сохранение cleaned-версии

```python
import os

df.to_csv(
    os.path.join(PROJECT_ROOT, "data", "cleaned_data.csv"),
    index=False
)
```
